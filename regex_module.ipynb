REgular Expression (REGEX)

```python
import re
```

```python
cumle = "Mustafa Kemal Atatürk, Türk asker, devlet adamı ve Türkiye Cumhuriyeti'nin kurucusudur."

patern = "Türk"

re.search(patern,cumle)
```

```python
durum = re.search(patern,cumle)
durum.span()
```

```python
dir(durum)
```

```python
durum.end()
```

```python
durum.group()
```

Çoklu Eşleşmelerde (match)

```python
for eslesme in re.findall(patern,cumle): #findall string döndürüyor
    print(eslesme)
```

```python
for eslesme in re.finditer(patern,cumle):
    print(eslesme.span(),eslesme.group())
```

Dinamik Kullanım

```python
ornek = "En sevdiğim kanal base42"
patern = r"base\d\d"
re.search(patern,ornek)
```

```python
cumle = "Selam, benim telefon numaram 0535-8886622"
patern = r"\d\d\d\d-\d\d\d\d\d\d\d\d"
re.search(patern,cumle)
```

```python
cumle = "Selam, benim telefon numaram 0535-8886622"
patern = r"\d{3,4}-\d{7}"
re.search(patern,cumle)
```

```python
cumle = "En sevdiğim kanal base42"
patern = r"\s\w{5,}"

for eslesme in re.finditer(patern, cumle):
    print(eslesme.group(),eslesme.span())
```

```python
patern = r"\d?"

for eslesme in re.finditer(patern, cumle):
    print(eslesme.group(),eslesme.span())
```

```python
patern = r"\w*\d+"

for eslesme in re.finditer(patern,cumle):
    print(eslesme.group(),eslesme.span)
```
