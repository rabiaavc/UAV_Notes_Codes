1-	Bir öğrenci not takip sistemi yazın. Program öğrenci bilgilerini (ad- soyad, vize notu, final notu) dictionary olarak saklasın. Tüm öğrencileri bir liste içinde tutsun. Ortalama hesaplayıp geçme/kalma durumunu belirlesin (vize %40, final %60, geçer not 50). Tüm öğrencileri ve durumlarını ekrana yazdırsın.

```python
students = [
    {"name": "Rabia",  "surname": "Avcı",   "midterm": 100, "final": 100},
    {"name": "Ali",    "surname": "Kaya",   "midterm": 60,  "final": 80},
    {"name": "Ahmet",  "surname": "Doğru",  "midterm": 70,  "final": 90},
    {"name": "Ece",    "surname": "Yılmaz", "midterm": 100, "final": 90},
    {"name": "Yağmur", "surname": "Kış",    "midterm": 100, "final": 90},
    {"name": "Hande",  "surname": "Demir",  "midterm": 50,  "final": 40},
    {"name": "Murat",  "surname": "Yıldız", "midterm": 30,  "final": 30},
]

for student in students:
    if (student["midterm"]*0.40 + student["final"]*0.60) >= 50:
        print("{} {} sınavdan geçti.".format(student["name"],student["surname"]))
    else:
        print("{} {} sınavdan kaldı.".format(student["name"],student["surname"]))
```

2-	Programınız “Python, Guido van Rossum tarafından 1991 yılında geliştirilmiş bir programlama dilidir. Python, okunabilirliği ve basit sözdizimi ile öne çıkar. Dilin tasarım felsefesi, kod okunabilirliğini vurgular ve bu da onu yeni başlayanlar için ideal kılar. Python, web geliştirme, veri analizi, yapay zeka, bilimsel hesaplama ve otomasyon gibi birçok alanda kullanılır. Python'un geniş kütüphane ekosistemi, geliştiricilere güçlü araçlar sunar. Python topluluğu çok aktiftir ve sürekli olarak dilin gelişimine katkıda bulunur. Python programlama dilini öğrenmek, yazılım dünyasında birçok kapı açar.” bu metinde en çok geçen 3 kelimeyi bulsun. Tüm harfleri küçük yapsın ve noktalama işaretlerini temizlesin.

```python
text = """Python, Guido van Rossum tarafından 1991 yılında geliştirilmiş bir programlama dilidir.
Python, okunabilirliği ve basit sözdizimi ile öne çıkar. Dilin tasarım felsefesi, kod okunabilirliğini vurgular ve bu da onu yeni başlayanlar için ideal kılar.
Python, web geliştirme, veri analizi, yapay zeka, bilimsel hesaplama ve otomasyon gibi birçok alanda kullanılır.
Python'un geniş kütüphane ekosistemi, geliştiricilere güçlü araçlar sunar. Python topluluğu çok aktiftir ve sürekli olarak dilin gelişimine katkıda bulunur.
Python programlama dilini öğrenmek, yazılım dünyasında birçok kapı açar."""
# En çok geçen 3 word
text_words = text.split()
punctuation = [",","."]
new_words = []
# noktalam işaretlerini temizleyelim ve küçük harf ile yazalım
for k in text_words:
    k = k.lower()
    length = len(k)
    if k[-1] == punctuation[0] or  k[-1] == punctuation[1] :
        k = k[:-1]
        new_words.append(k)
    else:
        new_words.append(k)

count = {}
for word in new_words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1

sorted(count.values(),reverse= True)

top3 = [] 
for i in range(3):
    max_value = 0
    max_word = ""

    for word in count:
        if word not in top3 and count[word] > max_value:
            max_value = count[word]
            max_word = word
    top3.append(max_word)

for word in top3:
    print(word, ":", count[word])
```

3-	Kullanıcı kayıt ve giriş sistemi oluşturun.

•	Kullanıcı adı en az 3 karakter olmalı.

•	 Şifre en az 6 karakter ve en az 1 rakam içermeli. 

•	E-mail’de “@” işareti olmalı. 

•	Yaş 13’ten büyük olmalı.

 Tüm koşullar sağlanıyorsa True, değilse False döndürsün.

```python
username = input("Enter username: ")
password = input("Enter password: ")
e_mail = input("Enter e-mail: ")
age = int(input("Enter your age: "))

boolean = True

if len(username) < 3:
    boolean = False
elif age < 13:
    boolean = False
elif len(password) < 6:
    boolean = False
else:
    has_digit = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
            break
    if not has_digit:
        boolean = False
    elif "@" not in e_mail:
        boolean = False
    else:
        boolean = True
print(boolean)
```

5-	Kullanıcının yaşına ve gelirine göre kredi başvurusu değerlendiren program yazın.

•	Yaş 18’den küçükse: “Yaşınız kredi için uygun değil.”

•	Yaş 18-65 arası ve gelir 5000’den fazla: “Kredi onaylandı.”

•	Yaş 18-65 arası ve gelir 5000’den az: “Geliriniz yetersiz.”

•	Yaş 65’ten büyük: “Yaşınız kredi için uygun değil.”

```python
yas = int(input("Yaşınızı girniz: "))
gelir = int(input("Gelirinizi giriniz: "))

if yas < 18:
    print("Yaşınız kredi için uygun değil.")
elif yas < 65 and gelir > 5000:
    print("Kredi onaylandı.")
elif yas < 65 and gelir < 5000:
    print("Geliriniz yetersiz.")
elif yas > 65:
    print("Yaşınız kredi için uygun değil.")
```

6-	Bir akıllı ev sisteminde, oturma odası sıcaklığı her saat başı ölçülüyor ve kaydediliyor. Başlangıç sıcaklıkları: (22.5, 23.1, 21.8, 19.5, 18.9, 20.2, 22.7, 24.3, 25.6, 26.2, 25.8, 24.9, 23.7, 22.4, 21.1, 20.8, 19.9, 18.7, 19.2, 20.5, 21.9, 23.3, 22.8, 21.5). 
İstenen işlemler:

•	Konfor Analizi: İdeal oda sıcaklığı 20-24 derece arasıdır. Bu aralıkta olmayan saatleri bulun.

•	Enerji Tüketim Uyarısı: 25 derece üzeri sıcaklıklarda klima, 18 derece altında ısıtıcı çalışıyor. İkisi için de ayrı ayrı çalışan saat sayısını bulun.

•	Gün içindeki en yüksek ve en düşük sıcaklık arasındaki farkı hesaplayın.

•	Günlük ortalama sıcaklığı hesaplayın.

```python
temp = [22.5, 23.1, 21.8, 19.5, 18.9, 20.2, 22.7, 24.3, 25.6, 26.2, 25.8, 24.9, 23.7, 22.4, 21.1, 20.8, 19.9, 18.7, 19.2, 20.5, 21.9, 23.3, 22.8, 21.5]
not_suitable = []
klima = []
Heater = []
sum = 0
for i in temp:
    sum += i
    if i < 20 or i > 24:
        not_suitable.append(i)
    if i > 25:
        klima.append(i)
    if i < 18:
        Heater.append(i)

sort_temp = sorted(temp)
avg = sum / len(temp)

print("20-24 derece arasında olmayan sıcaklıklar : {}".format(not_suitable))
print("Klima çalışan saat : {}\nIsıtıcı çalışan saat: {}".format(len(klima),len(Heater)))
print("En yüksek ve en düşüksıcaklık farkı : {}".format(sort_temp[-1] - sort_temp[0]))
print("Günlük ortalama sıcaklık : {}".format(avg))
```
