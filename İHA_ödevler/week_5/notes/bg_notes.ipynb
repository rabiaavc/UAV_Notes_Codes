```python
import cv2 as cv

cap = cv.VideoCapture(0)

background_sub = cv.createBackgroundSubtractorMOG2() #ışığa duyarlı
bs = cv.createBackgroundSubtractorKNN()

while True:
    ret,frame = cap.read()
    mask = bs.apply(frame)
    mask = cv.medianBlur(mask,5)


    cv.imshow("mask",mask)
    cv.imshow("frame",frame)
    if cv.waitKey(1) & 0xFF == ord("q"):
        break

cv.release()
cv.destroyAllWindows()
```

```python
#Erosion Aşındırma 
import cv2 as cv
import numpy as np

# Örnek için kamera veya maske kullanabilirsiniz
# Biz örnek olması için bir maske okuyalım
mask = cv.imread("mask.png", 0)  # 0 → grayscale (siyah-beyaz)

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
erosion = cv.erode(mask, kernel, iterations=1)

cv.imshow("Original Mask", mask)
cv.imshow("Erosion", erosion)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Dilation Genişleme
import cv2 as cv
import numpy as np

mask = cv.imread("mask.png", 0)  # siyah beyaz maske

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
dilation = cv.dilate(mask, kernel, iterations=1)

cv.imshow("Original Mask", mask)
cv.imshow("Dilation", dilation)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Birlikte Kullanımı 

mask = cv.imread("mask.png", 0)

kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))

# açma işlemi (temiz maske)
clean_mask = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel)

cv.imshow("Original Mask", mask)
cv.imshow("Clean Mask (Opening)", clean_mask)
cv.waitKey(0)
cv.destroyAllWindows()
```
