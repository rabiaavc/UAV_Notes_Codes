```python
import cv2 as cv
```

```python
path = r"cicek.jpg"
img = cv.imread(path)

type(img)
#pencerede tuttuk
cv.namedWindow("opencv_test", cv.WINDOW_AUTOSIZE)
#ekranda gstermek
cv.imshow("opencv_test",img)
cv.waitKey(1)
#tm pencereleri kapat
cv.destroyAllWindows()
```

```python
if img is None:
    print("Resim okunamad!")
else:
    print("shape:", img.shape)   # (ykseklik, genilik, kanal)

    h, w, c = img.shape
    print("Ykseklik:", h)
    print("Genilik:", w)
    print("Kanal says:", c)

mv = cv.split(img)
# Kanal1 = Blue
print(mv[0])
#Kanal2 = Green
print(mv[1])
#Kanal3 = Red
print(mv[2])
```

```python
b , g, r = cv.split(img)

#swap
k = r[:,:]
r[:,:] = b[:,:]
b[:,:] = k

swapped = cv.merge([b,g,r])

cv.imshow("Orijinal", img)
cv.imshow("Kanal Degistirilmis", swapped)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Grnty krpmak bu bir numpy dizisi olduu iin dizi stnde aritmetik ilemler yaplarak elde edilir
#ncelikle boyutlar yazyorum

h,w,c = img.shape

print(h,w,c)

#keden bir dikdtgen kesiyorum

crop = img[0:200, 0:300]

cv.imshow("crop", crop)
cv.waitKey(0)
cv.destroyAllWindows()

#yeni boyutlar
h1,w1,c1 = crop.shape

print(h1,w1,c1)
```

```python
#Yeniden Boyutlandrma  #cv.resize

new_size = (400,400)
resized = cv.resize(img,new_size)

cv.imshow("Yeniden Boyutlandrma",resized)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Gri Tona evirme cvtColor(dizin bilgisi,RGB KODU)

gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow("gray", gray)
cv.waitKey(3000)
```

```python
#zel Filtreler Uygulama
# applyColorMap
# eitli renk skalalar asndan resmi zenginletirmeye yarar

#COLORMAP_AUTUMN
#COLORMAP_BONE
#COLORMAP_WINTER
#COLORMAP_OCEAN
#COLORMAP_SUMMER
#COLORMAP_PINK
#COLORMAP_COOL
#COLORMAP_JET

dst = cv.applyColorMap(img,cv.COLORMAP_OCEAN)
cv.imshow("output",dst)
cv.waitKey(0)
cv.destroyAllWindows()
```
