```python
for i in range(1,11):
    for j in range(1,11):
        print("{} * {} = {}".format(i,j,(i*j)))
        if j == 10:
            print("----------------------")
```

3-	Bilgisayarın 1-100 arasında rastgele tuttuğu bir sayıyı kullanıcının tahmin etmeye çalıştığı bir program yazın. Kullanıcıya ipucu verin( daha büyük/daha küçük), maksimum 10 deneme hakkı tanıyın, doğru tahminde oyunu sonlandırın.

```python
import random

num = int(input("Enter a number between 1 and 100"))
predict = random.randint(1, 100)
count = 0
while count < 10:
    if num == predict:
        print("You are win")
        break
    elif num < predict:
        print("Your number is smaller than predict")
        num = int(input("Enter your new number"))
        continue
    elif num > predict:
        print("Your number is bigger than predict")
        num = int(input("Enter your new number"))
        continue
    count += 1
```

5-	Verilen 2 listeyi zip fonksiyonu ile birleştirerek index ve değerleri birlikte yazdırın.

[“Ali”, ”Ayşe”, ”Mehmet”, ”Zeynep”]

[25, 30, 35, 28]

```python
names = ["Ali", "Ayşe", "Mehmet", "Zeynep"]
numbers = [25, 30, 35, 28]
for index, name in zip(numbers,names):
    print("{} - {}".format(index,name))
```

6-	Range fonksiyonunu kullanarak;

a)	1’den 100’e kadar olan çift sayıların toplamını bulun.

b)	50’den geriye doğru 40’a kadar olan sayıları yazdırın.


```python
sum = 0
for i in range(1,100):
    if i % 2 == 0:
        sum += i
print("Even numbers of sum {}".format(sum))
```
